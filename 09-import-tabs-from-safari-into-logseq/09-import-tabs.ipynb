```python
# In this file, we will import the list of tabs from the frontmost Safari window.
# We will then activate Logseq and create a new node for each tab.
# The nodes will be created by putting the text into the clipboard,
# and then pasting it into Logseq.
print("go!")
import appscript
print("imported appscript")
# import library for setting the clipbord
import pyperclip
print("imported pyperclip")
# import clipboard
# import library for sending keystrokes
# import keyboard
# print("imported keyboard")
# import another library for sending keystrokes
import pyautogui
print("imported pyautogui")


from time import sleep

def url_and_title_for_each_tab_from_frontmost_safari_window():
    safari = appscript.app('Safari')
    res = []
    for tab in safari.windows[1].tabs():
        res.append({
            'url': tab.URL(),
            'title': tab.name()
        })
    return res

def activate_app_by_name(app_name):
    # if app is already activated, return
    if appscript.app('System Events').application_processes[app_name].frontmost.get():
        return

    appscript.app(app_name).activate()
    sleep(0.25)

def put_text_into_clipboard(text):
    pyperclip.copy(text)

def paste_text_from_clipboard():
    # use pyperclip to paste the text
    # pyperclip.paste()
    # use keyboard to paste the text
    # keyboard.send("command+v")
    # use pyautogui to paste the text
    pyautogui.hotkey('command', 'v')

def text_for_url_and_title(url, title):
    return f'[{title}]({url})'

def press_enter():
    pyautogui.press('enter')
    # keyboard.press_and_release('enter')
def press_tab():
    pyautogui.press('tab')
    # keyboard.press_and_release('tab')
def press_shift_tab():
    pyautogui.hotkey('shift', 'tab')
    # keyboard.press_and_release('shift+tab')
def type_text(text):
    pyautogui.typewrite(text)
    # keyboard.write(text)

def paste_text(text):
    put_text_into_clipboard(text)
    paste_text_from_clipboard()
def make_node_with_text(text):
    paste_text(text)
    press_enter()

def slash_command(command):
    # type "/"
    appscript.app('System Events').keystroke('/')
    # type the command keystrokes
    paste_text(command)
    sleep(0.05)
    # press enter
    press_enter()

# we will get and set a persistent window number variable
# import library for simple, easy persistent variables
import shelve

def get_window_number():
    with shelve.open('window_number') as db:
        return db.get('window_number', 1)
def set_window_number(window_number):
    with shelve.open('window_number') as db:
        db['window_number'] = window_number

import plumbum
# import colors
from plumbum import colors

window_number = get_window_number()
print(f'window_number: {window_number}')
set_window_number(window_number + 1)

sleep(1)

print("Activating Logseq...", end='')
activate_app_by_name('Logseq')
print(" done"&colors.green)

# formatted_date_and_time = ""
paste_text("Safari Window " + str(window_number) + " ")
slash_command("today")
slash_command("time")
press_enter()
press_tab()

res = url_and_title_for_each_tab_from_frontmost_safari_window()
# import pp
import pprint as pp
pp.pprint(res)


# iterate through the list of tabs
for tab in res:
    # get the url and title
    url = tab['url']
    title = tab['title']
    # make the text for the node
    text = text_for_url_and_title(url, title)
    # make the node
    make_node_with_text(text)

press_shift_tab()
```
