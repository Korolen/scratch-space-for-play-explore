```python
import plotly.express as px
# empty dataset
df = pd.DataFrame()
# line spiraling out from (0,0)

# import package np
import numpy as np

def line_spiral(n, a, b):
    return [(a + b * theta) * np.cos(theta) for theta in np.linspace(0, n * 2 * np.pi, 1000)], [(a + b * theta) * np.sin(theta) for theta in np.linspace(0, n * 2 * np.pi, 1000)]

spiral = line_spiral(1, 1, 0.1)
df['x'] = spiral[0]
df['y'] = spiral[1]
df['color'] = 'yo'

# plot
fig = px.line(df, x='x', y='y', color='color', title='Line Spiral')

# show plot
fig.show()
```
